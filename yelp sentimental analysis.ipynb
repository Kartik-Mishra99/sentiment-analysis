# Analysing reviews into positive or negitive or nuetral sentiments

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
def sentiment_value(paragraph):
    analyser = SentimentIntensityAnalyzer()
    result = analyser.polarity_scores(paragraph)
    score = result['compound']
    return round(score,1)
    
```

```python
sample = x[2001]
print(sample)
print('Sentiment: ')
print(sentiment_value(sample))
# sentiment score is 0.7
```

```python
sample = x[9008]
print(sample)
print('Sentiment: ')
print(sentiment_value(sample))
```

```python
all_sent_values = []
```

```python
for i in range(0,10000):
    all_sent_values.append(sentiment_value(x[i]))
```

```python
all_sent_values
```

```python
len(all_sent_values)
```

```python
SENTIMENT = []
SENTIMENT_VALUE = []
for i in range(0,10000):
    s = all_sent_values[i]
    if (s >=5 and s<=1):
        SENTIMENT.append('very_positive')
        SENTIMENT_VALUE.append(5)
    elif (s >0 and s < 5):
        SENTIMENT.append('positive')
        SENTIMENT_VALUE.append(4)
    elif (s == 0):
        SENTIMENT.append('neutral')
        SENTIMENT_VALUE.append(3)
    elif (s < 0 and s > -0.5):
        SENTIMENT.append('negitive')
        SENTIMENT_VALUE.append(2)
    else:
        SENTIMENT.append('very_negitive')
        SENTIMENT_VALUE.append(1)
        
```

```python
yelp['SENTIMENT'] = SENTIMENT
```

```python
yelp['SENTIMENT_VALUE'] = SENTIMENT_VALUE
```

# SENTIMENTAL ANALYSIS ON TRAINING AND TESTING DATA AND CHECKING ACCURACY

```python
yelp = pd.read_csv("C:/Users/Dell/Desktop/yelp.csv")
```

```python
yelp.keys()
```

```python
x = yelp['text']
y = yelp['stars']
```

```python
x
```

```python
y
```

```python
valid_revs = []
valid_labels = []
```

# as data is too large. so we take only those with labels value as 5 or 1  

```python
for i in range(x.shape[0]):
    if (y[i] == 5 or y[i] == 1):
        valid_revs.append(x[i])
        valid_labels.append(y[i])
        
out_revs = np.array(valid_revs)
print(out_revs.shape)

out_labels = np.array(valid_labels)
print(out_labels.shape)
        
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(out_revs,out_labels)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vector = CountVectorizer(stop_words='english',max_features=4000)
```

```python
print(count_vector)
```

```python
x_train_counts = count_vector.fit_transform(x_train)
x_test_counts = count_vector.fit_transform(x_test)
```

```python
print(x_train_counts.shape)
print(x_test_counts.shape)
```

```python
x_train_counts
```

```python
print(x_train_counts[1,9])
```

```python
print(y_train.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(C=5)
```

```python
clf.fit(x_train_counts.toarray(),y_train)
```

```python
y_pred = clf.predict(x_test_counts.toarray())
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tf_transformer = TfidfTransformer(use_idf=False).fit(x_train_counts)
x_train_tf = tf_transformer.transform(x_train_counts)
x_test_tf = tf_transformer.transform(x_test_counts)
```
